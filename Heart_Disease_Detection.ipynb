```python
df['Gender'] = df['Gender'].map({1: 'Male', 0: 'Female'})
```

```python
df['Condition']=df['Condition'].map({1:'Malignant', 0:'Benign'})
```

```python
df['Chest_Pain']=df['Chest_Pain'].map({0:'Typical Angina', 1:'Atypical Angina', 2:'No-Angina Pain', 3:'Asymptomatic'})
```

```python
co_matrix_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(20, 20))
sns.heatmap(co_matrix_df.corr(), annot=True, fmt='.2f')
```

Here's a summary of the correlations:

    1. Age is negatively correlated with Max_Heart_Rate (-0.39), which means that younger individuals tend to have higher maximum heart rates. It's positively correlated with Num_Major_Vessel (0.27) and ST_Slope (0.27).

    2. Chest_Pain is positively correlated with Exercise_Induced_Angina (0.31) and Max_Heart_Rate (0.31), indicating that chest pain is more likely to occur with exercise-induced angina and higher maximum heart rates. It's negatively correlated with Resting_ECG (-0.40).

    3. Resting_BP shows weak positive correlations with Age (0.27), Cholesterol (0.13), and Fasting_Blood_Sugar (0.18).

    4. Cholesterol has a weak positive correlation with Fasting_Blood_Sugar (0.18).

    5. Fasting_Blood_Sugar has a weak positive correlation with Cholesterol (0.18).

    6. Resting_ECG is negatively correlated with Chest_Pain (-0.40) and Exercise_Induced_Angina (-0.38), and has a weak positive correlation with Age (0.13).

    7. Max_Heart_Rate is negatively correlated with Age (-0.39) and Exercise_Induced_Angina (-0.35), and positively correlated with Chest_Pain (0.31).

    8. Exercise_Induced_Angina is positively correlated with Chest_Pain (0.31) and negatively correlated with Max_Heart_Rate (-0.35).

    9. ST_Depression is negatively correlated with Max_Heart_Rate (-0.35) and ST_Slope (-0.58), and positively correlated with Exercise_Induced_Angina (0.31).

    10. ST_Slope is positively correlated with Max_Heart_Rate (0.40) and negatively correlated with ST_Depression (-0.58).

    11. Num_Major_Vessel is positively correlated with Age (0.27) and Chest_Pain (0.27).
    
    12. Thallium_Test does not show any strong correlations with other features.

Key Findings from the Density Plot:

    1. Both benign and malignant conditions show peaks around similar ages (45 and 55).
    2.The malignant condition curve shows a higher density around the first peak (age 45) compared to the benign curve, suggesting that younger individuals in their 40s might be more prone to malignant conditions compared to benign ones.
