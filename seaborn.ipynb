```python
fields=['AI','ML','DS','DL','DJ']
div_a=[12,34,23,16,20]
div_b=[10,20,22,13,11]
div_c=[13,25,26,15,14]
index=np.arange(len(fields)*2, step=2)
```

```python
fig,ax=plt.subplots()
bar_width = 0.5
bars=plt.bar(index-bar_width,div_a, width=0.5, color='#f2cc8f',label='DIV_A')
bars1=plt.bar(index ,div_b, width=0.5, color='#84a98c', label='DIV_B')
bars2=plt.bar(index + bar_width,div_c, width=0.5, color='#4a4e69',  label='DIV_C')
ax.bar_label(bars)
ax.bar_label(bars1)
ax.bar_label(bars2)
plt.xticks(index , fields)
plt.title('FIELD-WISE INTERESTED STUDENTS', fontsize=15)
plt.tight_layout()
plt.xlabel('FIELDS', fontsize=12)
plt.ylabel('STUDENTS', fontsize=12)
plt.legend()
plt.show()
```

```python
population=np.arange(5,50,5)
population
```

```python
pop_mean=sum(population)/len(population)
```

```python
pop_mean
```

```python
sq_pop_deviation=[(i-pop_mean)**2 for i in population]
sq_pop_deviation  #sq_pop_deviation= square population deviation
```

```python
pop_variance=sum(sq_pop_deviation)/len(population)
```

```python
pop_variance
```

```python
import math
std_dev=math.sqrt(pop_variance)
print(f"Standard Population Devation: {round(std_dev,2)}")
```

```python
sample=np.arange(5,25,5)
```

```python
sample
```

```python
s_mean=sum(sample)/len(sample)
```

```python
s_mean
```

```python
sq_s_deviation=[(i-s_mean)**2 for i in sample]
sq_s_deviation
```

```python
s_variance=sum(sq_s_deviation)/(len(sample)-1)
```

```python
print(f"Sample Variance: {round(s_variance,1)}")
```

```python
std_s_dev=math.sqrt(s_variance)
print(f"Standard Sample Devation: {round(std_s_dev,2)}")
```
